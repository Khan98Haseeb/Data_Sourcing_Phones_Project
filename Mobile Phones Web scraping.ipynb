```python
# Import useful libraries and classes.
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
```

```python
my_url = 'https://www.91mobiles.com/top-10-mobiles-in-india'
uClient = uReq(my_url)
page_html = uClient.read()
uClient.close()
```

```python
type(page_html)
```

```python
page_html
```

```python
page_soup = soup(page_html, 'html.parser')
page_soup
```

```python
containers= page_soup.findAll("div", {"class": "filter filer_finder"})
print(len(containers))
```

```python
filename= "/content/drive/MyDrive/UpGrad Data Science/The Course/Course 1 - Data Toolkit/Exploratory Data Analysis/Mobile_Phones_EDA.csv"
f= open(filename, "w")
```

```python
headers= "Name, Processor, Display, Camera \n"
f.write(headers)
```

```python
print(soup.prettify(containers[0]))
```

```python
# headers= "Name, Processor, Display, Camera \n"
for container in containers:
    name= container.a['title']
    feats_phone= container.findAll("div", {"class": "left specs_li"})
    pross = feats_phone[0].findAll('label', {'class':' '})[0]['title']
    pross = pross.replace(',', '')
    disp = feats_phone[1].findAll('label', {'class':' '})[0]['title']
    cam = feats_phone[2].findAll('label', {'class':' '})[0]['title']

    # print(name)
    # print(pross)
    # print(disp)
    # print(cam)
    print(name + "," + pross + "," + disp + "," + cam + "\n")
    f.write(name + "," + pross + "," + disp + "," + cam + "\n")
    
f.close()
```
